```python
import json 

with open("./data/countyGeoJson.js") as f:
    geodata = json.load(f)

with open("./data/combined_data.js") as f:
    data = json.load(f)
    
```

```python
for x in data:
    for y in geodata["features"]:
        if y["properties"]["name"] == x["county"]:
            y["properties"]["data"] = x

countyGeoData = geodata
countyGeoData
```

```python
with open('countyGeoData.js', 'w') as fp:
    json.dump(countyGeoData, fp, sort_keys=False, indent=4)
```
